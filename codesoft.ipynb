```python
def simple_chatbot(user_input):
    greetings = ["hello", "hi", "hey", "greetings"]
    farewells = ["bye", "goodbye", "see you", "take care"]
    Queries=["What is your name","tell me your name"]
    jokes=[ "make me laugh","tell joke"]
    Query2=["How are you","what about you"]


    if any(word in user_input.lower() for word in greetings):
        return "Hello! How can I help you today?"

    # Check if the user's input is a farewell
    elif any(word in user_input.lower() for word in farewells):
        return "Goodbye! Have a great day."

    elif any(word in user_input.lower() for word in jokes):
        return "I went to a seafood disco last week and pulled a mussel."

    elif any(word in user_input.lower() for word in Queries):
        return "You can call me chatbot."

    elif any(word in user_input.lower() for word in Query2):
        return "I am fine and i am doing good!."


    else:
        return "I'm just a simple chatbot. I don't understand that."


print("SimpleBot: Hello! Type 'exit' to end the conversation.")
while True:


    user_input = input("You: ")


    if user_input.lower() == 'exit':
        print("SimpleBot: Goodbye!")
        break


    response = simple_chatbot(user_input)
    print("SimpleBot:",response)
```

```python
import random

# Constants for player symbols
HUMAN_PLAYER = 'X'
AI_PLAYER = 'O'
EMPTY = ' '

def print_board(board):
    for row in board:
        print("|".join(row))
        print("-----")

def is_winner(board, player):
    # Check rows, columns, and diagonals for a win
    return any(all(cell == player for cell in row) for row in board) or \
           any(all(board[i][j] == player for i in range(3)) for j in range(3)) or \
           all(board[i][i] == player for i in range(3)) or \
           all(board[i][2 - i] == player for i in range(3))

def is_board_full(board):
    return all(cell != EMPTY for row in board for cell in row)

def get_empty_cells(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY]

def minimax(board, depth, maximizing_player):
    if is_winner(board, HUMAN_PLAYER):
        return -1
    elif is_winner(board, AI_PLAYER):
        return 1
    elif is_board_full(board):
        return 0

    if maximizing_player:
        max_eval = float('-inf')
        for i, j in get_empty_cells(board):
            board[i][j] = AI_PLAYER
            eval = minimax(board, depth + 1, False)
            board[i][j] = EMPTY
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for i, j in get_empty_cells(board):
            board[i][j] = HUMAN_PLAYER
            eval = minimax(board, depth + 1, True)
            board[i][j] = EMPTY
            min_eval = min(min_eval, eval)
        return min_eval

def find_best_move(board):
    best_val = float('-inf')
    best_move = None
    for i, j in get_empty_cells(board):
        board[i][j] = AI_PLAYER
        move_val = minimax(board, 0, False)
        board[i][j] = EMPTY

        if move_val > best_val:
            best_move = (i, j)
            best_val = move_val

    return best_move

def play_game():
    board = [[EMPTY] * 3 for _ in range(3)]
    print_board(board)

    while True:
        # Human player's turn
        row, col = map(int, input("Enter your move (row and column, separated by space): ").split())
        if board[row][col] != EMPTY:
            print("Cell already occupied. Try again.")
            continue
        board[row][col] = HUMAN_PLAYER

        print_board(board)

        if is_winner(board, HUMAN_PLAYER):
            print("Congratulations! You win!")
            break
        elif is_board_full(board):
            print("It's a draw!")
            break

        # AI player's turn
        print("AI's move:")
        ai_row, ai_col = find_best_move(board)
        board[ai_row][ai_col] = AI_PLAYER

        print_board(board)

        if is_winner(board, AI_PLAYER):
            print("AI wins! Better luck next time.")
            break
        elif is_board_full(board):
            print("It's a draw!")
            break

if __name__ == "__main__":
    play_game()
```
